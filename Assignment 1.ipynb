```python
# import networkx library
import networkx as nx

# create nodes
nodes = ["Saayed", "Anan", "Shayaan", "Dayan", "Shathi", 
         "Shawon", "Hasna", "Lily", "Dulal", "Ferdous"]

# create edges
edges = [("Saayed", "Anan"), 
         ("Saayed", "Shayaan"),
         ("Saayed", "Shathi"),
         ("Saayed", "Shawon"),
         ("Saayed", "Lily"),
         ("Saayed", "Dulal"), 
         ("Shayaan", "Shathi"), 
         ("Shathi", "Shawon"), 
         ("Shawon", "Lily"), 
         ("Lily", "Dulal"), 
         ("Dulal", "Anan"), 
         ("Anan", "Shayaan"), 
         ("Anan", "Dayan"), 
         ("Dayan", "Shayaan"), 
         ("Dayan", "Hasna"), 
         ("Hasna", "Ferdous"), 
         ("Shayaan", "Shawon"), 
         ("Anan", "Lily")]

# draw a basic graph using nx 
G = nx.Graph()

# add nodes to the graph
G.add_nodes_from(nodes)

# add edges to the graph
G.add_edges_from(edges)

# draw graph
nx.draw(G, 
        with_labels = True, font_weight = "bold", 
        node_color = "#4286f4", alpha = 0.5, node_shape = "d")


```
